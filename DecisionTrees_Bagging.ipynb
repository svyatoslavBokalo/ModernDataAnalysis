#### Function to check if the stock outperforms SPY by more than 1% in the earnings period, then return 1. If it underperforms by -2.5% or less, then return -1, otherwise, return 0.

```python
def f(x):
    if x > 0.01:
        return 1
    elif x <= -0.025:
        return -1
    else:
        return 0
```

## Use Optuna to find optimal hyperparameters.

```python
#!pip install optuna
```

```python
import optuna
from optuna.trial import Trial
from functools import partial
```

```python
def objective(trial:Trial,
              train=None,
              labels=None,
              val=None,
              val_labels=None,
              val_rets=None):

    t_min_samples_leaf = trial.suggest_int('min_samples_leaf', 100, 1200, step=100)
    t_max_depth = trial.suggest_int('max_depth', 5, 25, step=5)

    
    t_clf = DecisionTreeClassifier(min_samples_leaf = t_min_samples_leaf, 
                                   max_depth = t_max_depth, 
                                   random_state = 123)
    t_clf.fit(train, labels)

    preds = t_clf.predict(val)
    profit = (preds * val_rets).sum()

    return profit
```

```python
study = optuna.create_study(direction="maximize")
```

```python
%%time
study.optimize(partial(objective, train=train, labels=y_train, val=valid, val_labels=y_valid, val_rets=valid_stock_returns.values), n_trials=20, n_jobs=-1)
```

```python
study.best_params
```

### Instantiating the DecisionTreeClassifier with hyper-parameters and train the model

```python
t_clf = DecisionTreeClassifier(**study.best_params, random_state=123)
t_clf.fit(train,y_train)
```

### Create the objective for BoostingClassifier

```python
def objective_b(trial:Trial,
              train=None,
              labels=None,
              val=None,
              val_labels=None,
              val_rets=None):

    b_n_estimators = trial.suggest_int('n_estimators', 100, 1000, step=100)
    b_max_samples = trial.suggest_int('max_samples', 50, 1000, step=50)

    
    b_clf = BaggingClassifier(estimator = DecisionTreeClassifier(),
                          n_estimators = b_n_estimators,
                          max_samples = b_max_samples,
                         random_state = 42)
    b_clf.fit(train, labels)

    preds = b_clf.predict(val)
    profit = (preds * val_rets).sum()

    return profit
```
